```python
#Load env variables and create client
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
#Helper functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

#Chat function without system prompt
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text

#Chat function with system prompt
def chat(messages):
    system = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """

    message = client.messages.create(
        model=model,
        messages=messages,
        max_tokens=1000,
        system=system
    )
    return message.content[0].text

#To avoid hardcoding the system prompt in the chat function, you can also add it to the messages list as a system message. This way, you can easily modify or remove the system prompt without changing the chat function itself.
def chat(messages, system=None):
    message = client.messages.create(
        model=model,
        messages=messages,
        max_tokens=1000,
        system=system
    )
    return message.content[0].text

#If decided not to include the system prompt in the chat function, you can also add it to the messages list as a system message. This way, you can easily modify or remove the system prompt without changing the chat function itself.
def chat(messages, system=None):
    params = {
        "model" : model,
        "messages" : messages,
        "max_tokens" : 1000,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text

#Using Temperature to control the randomness of the model's output. A higher temperature will result in more random output, while a lower temperature will result in more focused and deterministic output. You can experiment with different temperature values to see how it affects the model's responses.
def chat(messages, system=None, temperature=1.0):
    params = {
        "model" : model,
        "messages" : messages,
        "max_tokens" : 1000,
        "temperature" : temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
#Claude response without a system message prompt
#You can re run the code below to see how the response changes with system prompts and without them.
messages = []
add_user_message(messages, "How do i solve 5x+3=2 for x?")
answer = chat(messages)

answer
```

```python
#reduce the amount of code that's being generated by the model by adding a system prompt that instructs the model to be concise in its response. This can help to reduce the number of tokens generated and make the response more focused.
messages = []
add_user_message(messages, "Write a Python function that checks a string for duplicate characters.")
answer = chat(messages, system="You are a Python engineer who writes very concise code")
answer
```

```python
# Temperature controls output randomness. 
# Use low values for strict, predictable tasks and high values for creative tasks. 
# Finding the right balance is key, as setting it too high can lead to irrelevant or nonsensical output.
messages = []
add_user_message(messages, "Generate a one sentence movie idea")
answer = chat(messages, temperature=0.0)
answer
```
